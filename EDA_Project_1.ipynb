<h3 style = "text-align:center;color:#00A86B">Asking questions and finding answers in the dataset</h3>

Some EDA questions that can be explored with the given dataset:

1. **General Statistics:**
    - What is the distribution of event distances (e.g., 50km, 50mi)?
    - What is the distribution of athlete genders across different event distances?

2. **Performance Analysis:**
    - What is the average performance time for different event distances?
    - How does the average performance time vary by gender for different event distances?

3. **Age Analysis:**
    - What is the distribution of athlete ages across different event distances?
    - How does athlete performance vary with age for different event distances?
    - Is there a significant difference in performance between different age categories?

4. **Country Analysis:**
    - Which countries have the most participants in the dataset?
    - How does the average performance time vary by country?

5. **Yearly Trends:**
    - How has the number of events and participants changed over the years?
    - Are there any noticeable trends in performance times over the years?

6. **Gender Comparison:**
    - How does the performance of male and female athletes compare across different event distances?
    - Are there any significant differences in the average speed of male and female athletes?

7. **Event-Specific Analysis:**
    - Which events have the highest number of finishers?
    - What are the top events in terms of average athlete performance?

8. **Speed Analysis:**
    - What is the distribution of average speeds for different event distances?
    - How does the average speed vary by gender and age?

9. **Club Analysis:**
    - Which athlete clubs have the most participants?
    - How does the performance of athletes from different clubs compare?

10. **Missing Data:**
     - How much missing data is there in each column?
     - How does the presence of missing data affect the analysis?

These questions can help guide your exploratory data analysis and uncover interesting insights from the dataset.
