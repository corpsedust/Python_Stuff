<h2> Linked List </h2>

Arrays have a disadvantage when you have to insert an element in betweeen. Here, say  you want to insert element at 3 you will have to shift the entire array to right so that 3 is empty. This is fine if you have small data set but it will be cumbersome if you have a huge dataset, not vert efficient.

### Singly Linked List
This contains a long chain of nodes. Each node stores some data that we need to store and the address to the next node. They do not have index like Arrays do instead they contain address where the next node is. Each individual node can be anywhere in the computer's memory

Picture this as a treasure hunt where each node contains the address of the next, so once you find a node you can find the next. When you reach a node that has no address, that's when you know you have reached the end. 



Now it's easy to insert or delete shit. Say you want to insert a node at some point, all you have to do is change address in the node before to address of new node. Then in the new node add the address of next node which was already in the node before. Less steps, no shifting.
Similarly you can do deleting a node, just change address in the previous node to point at node which is next to node which you wnat to be deleted.



### Doubly Linked List
Here each node has two addresses, one being of the next and the other being of the previous node. The advantage of this node over the previous one is that you can traverse this node from head to tail or tail to head. 

<font color = "red">There is disadvantage with list as compared to array, that is since there are no indices and if you have to find an element you have to traverse the entire list which adds to time complexity</font>




```python
#Singly Linked Lists


class SinglyNode:
    
    def __init__(self, val, next = None):
        self.val = val
        self.next = next
        
    def __str__(self):
        return str(self.val)


Head = SinglyNode(1)
A = SinglyNode(2)
B = SinglyNode(3)
C = SinglyNode(4)

Head.next = A
A.next = B
B.next = C

```

```python
#Traverse the list - O(n)

curr = Head

while curr:     #(goes on till the value of curr is none)
    print(curr)
    curr = curr.next
    
```

```python
#Display linked list - O(n)

def display(head):
    curr = head
    elements = []
    while curr:
        elements.append(str(curr.val))
        curr = curr.next
    print(' ->  '.join(elements))
    
display(Head)
```

```python
# Search for a node value = O(n)

def search(head, val):
    curr = head
    while curr:
        if val == curr.val:
            return True
        curr = curr.next
        
    return False

search(Head, 4)
```

### The big O notation 

"How code slows as the data grows"

1. Describes the performance of an algorithm as the amount of data increases.
2. Machine independednt (# of steps to completion)
3. Ignore smaller operations O(n+1) -> O(n)

* Example
    * O(1)
    * O(n)
    * O(log n)
    * O(n^2)
        
n = amount of data (it's a variable like x)

```python
#O(n) - linear time

def addup(n):
    s = 0
    for i in range(n+1):  #range by default starts from 0, and does not include the last number
        s += i
    return s

addup(4)

# this is a linear time complexity function, as the time taken to 
# execute the function is directly proportional to the input size
# that is, as the size of n increases, the numbers of steps increases linearly


#O(1) - constant time   

def addup2(m):
    return m*(m+1)/2

addup2(4)

# this will only take one step to execute, no matter the size of the input
 
 #addup2 is better function than addup, as it is more efficient
```

Given a sorted array of distinct integers and a target value, return the index if the target is found. 
If not, return the index where it would be if it were inserted in order.

```python
def binary_search(arr,x): #x is the number to be searched
    left , right  = 0, len(arr) - 1
    
    
    while left <= right:
        mid = (left+right) // 2

        if x == arr[mid]:
            return mid
        if x > arr[mid]:
            left = mid
        else:
            right = mid
    
    return left       
        
        


a1 = [45, 87, 89, 90, 92, 93, 95, 96, 97, 98, 99, 100]
x = 88
binary_search(a1,x)
```

<font color ="cyan"> Problem no. 3417 (Zigzag Traversal, Easy) </font>

You are given an m x n 2D array grid of positive integers.

Your task is to traverse grid in a zigzag pattern while skipping every alternate cell.

Zigzag pattern traversal is defined as following the below actions:

    Start at the top-left cell (0, 0).
    Move right within a row until the end of the row is reached.
    Drop down to the next row, then traverse left until the beginning of the row is reached.
    Continue alternating between right and left traversal until every row has been traversed.

Note that you must skip every alternate cell during the traversal.

Return an array of integers result containing, in order, the value of the cells visited during the zigzag traversal with skip

```python
#Initial solution

def zigzagTraversal(self, grid):
        """
        :type grid: List[List[int]]
        :rtype: List[int]
        """
        l1 = len(grid)      #length of the entire array
        l2 = len(grid[0])   #length of each individual list in the grid 
        a = []

        for i in range(l1):
            c = 0
            if i%2 == 0:
                for c in range(0,l2,2):
                    a.append(grid[i][c])
                    
            else:
                for c in range(1,l2,2):
                    a.append(grid[i][c])
                
                    
        return a
    
    
#Complexity is O(n,m) n is lengths of list containing list
#m is length of each list
#also one thing wrong with this solution is that when you go to odd index row
#your pointer should start from back - see the question on leetcode to understand
```

```python
#Internet Solution

def zigzagTraversal(self, grid):
        m, n = len(grid), len(grid[0]) 
        result = []
        
        
        for i in range(m):
            if i % 2 == 0: 
                row = grid[i][::2]  
            #when row index is even, simply take every second elements from indices 0,2,4...        
            else:  
                row = grid[i][::-1][n%2::2]   #reverse the array, from reversed array start at n%2 (could be one) and pick every second element
            #when row index is odd
            result.extend(row)
        
        return result
```
