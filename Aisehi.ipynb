```python
def stringsplicer(s):
    s2 = ""
    for i in s:
        if s2.find(i) == -1:
            s2 = s2 + i
        else:
            break;
    return s2

stringsplicer("abcabcbb")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dd = pd.read_csv("Nat_Gas.csv", parse_dates=True)
dd.set_index("Dates", inplace=True)
```

```python
dd.head()
plt.figure(figsize=(16,8))
plt.xlabel("Year")
plt.ylabel("Prices")
plt.plot(dd)
plt.show()
```

```python
from statsmodels.tsa.api import SimpleExpSmoothing
model = SimpleExpSmoothing(dd)
model_single_fit = model.fit()
forecast_single = model_single_fit.forecast(12)
ff = pd.DataFrame(forecast_single)
ff

```

```python
gg["Fitted Values"] = pd.DataFrame(model_single_fit.fittedvalues)
gg["OG"]  = dd['Prices']

gg
```

```python
plt.plot(dd, label = "Original Data")
plt.plot(model_single_fit.fittedvalues, label = "Fitted Values")
plt.plot(forecast_single, label = "Forecast")
plt.xlabel("Year")
plt.ylabel("Prices")
plt.title("Single Exponential Smoothing")
plt.legend()
plt.show()
```
